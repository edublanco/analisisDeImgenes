```python
dim = (400, 400)

images_dir = Path('img').expanduser()

# Resizing all the images to same dimension
image_train = []
for fname in listdir(images_dir):
    fpath = os.path.join(images_dir, fname)
    im = Image.open(fpath)
    im_resized = im.resize(dim)
    image_train.append(im_resized)


```

```python
image_train[2]
```

```python
## Converting the image to numpy array
train_x=[]
for x in range(len(image_train)):
    x_image=np.array(image_train[x],dtype='uint8')
    train_x.append(x_image)

train_x = np.asarray(train_x)
```
